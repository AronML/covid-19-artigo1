```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('arquivo_geral.csv', delimiter = ';')
```

```python
data
```

```python
data_Norte = data['regiao'] == 'Norte'
data_Norte = data[data_Norte]

data_AL = data['estado'] == 'AL'
data_AL = data[data_AL]

data_RO = data['estado'] == 'RO'
data_RO = data[data_RO]

data_AC = data['estado'] == 'AC'
data_AC = data[data_AC]

data_AM = data['estado'] == 'AM'
data_AM = data[data_AM]

data_RR = data['estado'] == 'RR'
data_RR = data[data_RR]

data_PA = data['estado'] == 'PA'
data_PA = data[data_PA]

data_AP = data['estado'] == 'AP'
data_AP = data[data_AP]

data_TO = data['estado'] == 'TO'
data_TO = data[data_TO]

data_MA = data['estado'] == 'MA'
data_MA = data[data_MA]

data_PI = data['estado'] == 'PI'
data_PI = data[data_PI]

data_CE = data['estado'] == 'CE'
data_CE = data[data_CE]

data_RN = data['estado'] == 'RN'
data_RN = data[data_RN]

data_PB = data['estado'] == 'PB'
data_PB = data[data_PB]

data_PE = data['estado'] == 'PE'
data_PE = data[data_PE]

data_SE = data['estado'] == 'SE'
data_SE = data[data_SE]

data_BA = data['estado'] == 'BA'
data_BA = data[data_BA]

data_MG = data['estado'] == 'MG'
data_MG = data[data_MG]

data_ES = data['estado'] == 'ES'
data_ES = data[data_ES]

data_RJ = data['estado'] == 'RJ'
data_RJ = data[data_RJ]

data_SP = data['estado'] == 'SP'
data_SP = data[data_SP]

data_PR = data['estado'] == 'PR'
data_PR = data[data_PR]

data_SC = data['estado'] == 'SC'
data_SC = data[data_SC]

data_RS = data['estado'] == 'RS'
data_RS = data[data_RS]

data_MS = data['estado'] == 'MS'
data_MS = data[data_MS]

data_MT = data['estado'] == 'MT'
data_MT = data[data_MT]

data_GO = data['estado'] == 'GO'
data_GO = data[data_GO]

data_DF = data['estado'] == 'DF'
data_DF = data[data_DF]
```

```python
plt.plot(data_MA['data'],data_MA['casosAcumulados'])
plt.plot(data_PI['data'],data_PI['casosAcumulados'])
plt.plot(data_CE['data'],data_CE['casosAcumulados'])
plt.plot(data_RN['data'],data_RN['casosAcumulados'])
plt.plot(data_PB['data'],data_PB['casosAcumulados'])
plt.plot(data_PE['data'],data_PE['casosAcumulados'])
plt.plot(data_AL['data'],data_AL['casosAcumulados'])
plt.plot(data_SE['data'],data_SE['casosAcumulados'])
plt.plot(data_BA['data'],data_BA['casosAcumulados'])
plt.xticks(['30/01/2020','10/02/2020'
            ,'20/02/2020','01/03/2020','10/03/2020','20/03/2020','30/03/2020',
            '10/04/2020','20/04/2020','30/04/2020', '04/05/2020'], rotation = 90)
plt.title('Total de casos - Região Nordeste')
plt.legend(['MA','PI','CE','RN','PB','PE','AL','SE','BA'])
plt.grid()
plt.savefig('total_casos_nordeste.png',bbox_inches='tight')
plt.show()
```
