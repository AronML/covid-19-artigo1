```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('covid19br.csv')
```

```python
data.recovered.fillna(0, inplace=True)
data.suspects.fillna(0, inplace=True)
data.tests.fillna(0, inplace=True)
data.tests_per_100k_inhabitants.fillna(0, inplace=True)
data
```

```python
plt.bar(data['state'], data['tests_per_100k_inhabitants'],edgecolor = 'black')
plt.bar(data['state'], data['totalCases_per_100k_inhabitants'],edgecolor = 'black')

plt.xticks(rotation = 'vertical')
plt.title('COVID-19 por 100 mil habitantes')
plt.legend(['testes','total de casos'])
plt.savefig('testes_casos_100k.png',bbox_inches='tight')
plt.show()

```

```python
plt.bar(data['state'], data['totalCases_per_100k_inhabitants'],edgecolor = 'black')
plt.bar(data['state'], data['deaths_per_100k_inhabitants'],edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('COVID-19 por 100 mil habitantes')
plt.legend(['total de casos','mortes'])

plt.savefig('casos_mortes_100k.png',bbox_inches='tight')

plt.show()
```

```python
plt.bar(data['state'], data['totalCases_per_100k_inhabitants'],edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('total de casos COVID-19 por 100 mil habitantes')
plt.grid()

plt.savefig('casos_100k.png',bbox_inches='tight')
plt.show()
```

```python
plt.bar(data['state'], data['deaths_per_100k_inhabitants'],edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('mortos por 100 mil habitantes')
plt.grid()

plt.savefig('mortos_100k.png',bbox_inches='tight')
plt.show()
```

```python
data_sem_total = data['state'] != 'TOTAL'
data_sem_total = data[data_sem_total]
data_sem_total
```

```python
plt.bar(data_sem_total['state'], data_sem_total['totalCases'],edgecolor = 'black')
plt.bar(data_sem_total['state'], data_sem_total['deaths'],edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('total de casos')
plt.legend(['total','mortes totais'])
plt.grid()
plt.savefig('casos_mortes.png',bbox_inches='tight')
plt.show()
```

```python

plt.bar(data['state'], data['deaths_by_totalCases']*100,edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('taxa de mortalidade(%)')
plt.grid()

plt.savefig('taxa_mortalidade.png',bbox_inches='tight')
plt.show()
```

```python
mortalidade_total = data['state'] == 'TOTAL'
mortalidade_total = (data[mortalidade_total].deaths_by_totalCases)*100
mortalidade_total
```

```python
mortalidade_min = min(data['deaths_by_totalCases'])*100
mortalidade_min
#mortalidade_min = (data[mortalidade_min].deaths_by_totalCases)*100
```

```python
correcao = mortalidade_total/mortalidade_min
#correcao = mortalidade_total/1.867
nMortes_total = data['state'] == 'TOTAL'
nMortes_total = data[nMortes_total].totalCases
possivel_total_casos = correcao*nMortes_total
possivel_total_casos
#nMortes_total
```

```python
plt.bar(data_sem_total['state'], data_sem_total['totalCases']+data_sem_total['suspects'],edgecolor = 'black')
plt.bar(data_sem_total['state'], data_sem_total['totalCases'],edgecolor = 'black')
plt.bar(data_sem_total['state'], data_sem_total['recovered'],edgecolor = 'black')
plt.xticks(rotation = 'vertical')
plt.title('recuperados')
plt.legend(['Suspeitos adicionais','total de casos','recuperados'])
plt.grid()

plt.savefig('suspeitos_recuperados.png',bbox_inches='tight')
plt.show()
```
